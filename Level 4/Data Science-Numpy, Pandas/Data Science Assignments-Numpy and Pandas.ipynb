#### Using the BicycleWeather.csv dataset perform the following operations: 
    1. Extract the PRCP column from the pandas object.
    2. Using boolean functions/operators separate numbers from NAN 
    3. Using boolean masking separate all zeros and non-zeros into different 
    variables. 
    4. From the variable holding non-zeros, multiply every values in the array with  
    100 and divid by 250 
    5. Calculate the following: 
        a. Median precip on rainy days 
        b. Median precip on summer days 
        c. Maximum precip on summer days 
        d. Median precip on non-summer rainy days

```python
#Load the dataset
bicycle_weather = pd.read_csv(r"C:\users\atade\Data Science and AI\Datasets\BicycleWeather.csv")
bicycle_weather.head(200)
```

```python
#Extract the PRCP column from the pandas object.
prcp = bicycle_weather['PRCP']
prcp
```

```python
#Using boolean functions/operators separate numbers from NAN 
numbers = prcp[~np.isnan(prcp)]
nan_values = prcp[np.isnan(prcp)]
numbers, nan_values
```

```python
#Using boolean masking separate all zeros and non-zeros into different variables. 
zeros = prcp[prcp == 0]
non_zeros = prcp[prcp != 0]
zeros, non_zeros
```

```python
#From the variable holding non-zeros, multiply every values in the array with 100 and divid by 250 
modified_non_zeros = non_zeros * 100 / 250
modified_non_zeros
```

```python
prcp1 = bicycle_weather['PRCP'].values
date = pd.to_datetime(bicycle_weather['DATE'], format='%Y%m%d')
rainy_days = prcp1 > 0  # Boolean mask for rainy days

# Boolean mask for summer months (June, July, August)
summer_days = date.dt.month.isin([6, 7, 8])

# Boolean mask for non-summer rainy days
non_summer_rainy_days = rainy_days & ~summer_days

# Calculating statistics
median_precip_rainy_days = np.median(prcp1[rainy_days])
median_precip_summer_days = np.median(prcp1[summer_days])
max_precip_summer_days = np.max(prcp1[summer_days])
median_precip_non_summer_rainy_days = np.median(prcp1[non_summer_rainy_days])

# Print results
print(f"Median precip on rainy days: {median_precip_rainy_days}")
print(f"Median precip on summer days: {median_precip_summer_days}")
print(f"Maximum precip on summer days: {max_precip_summer_days}")
print(f"Median precip on non-summer rainy days: {median_precip_non_summer_rainy_days}")
```

Using the datasets above: List_of_local_governments_in_Nigeria.csv and Nigeria_povertyrate_per_local_gov.csv

    1. Marge the two datasets together to form a single table.
    2. Remove any redundant columns
    3. Clean the dataset to remove every null values
    4. Using either groupby or pivort_table calculate how population affects poverty rate of each local government
    5. Arrange each local goverment according to the order of their poverty rate

```python
import pandas as pd
import numpy as np
# Load datasets
local_gov = pd.read_csv(r"C:\users\atade\Data Science and AI\Datasets\Nigeria data\list_of_local_government_areas_of_nigeria.csv")
poverty_rate = pd.read_csv(r"C:\users\atade\Data Science and AI\Datasets\Nigeria data\Nigeria povertyrate per local gov.csv")
local_gov.head()
poverty_rate.head()
```

```python
poverty_rate['SNo'] = range(1, len(poverty_rate) + 1)
poverty_rate.head()
```

```python
poverty_rate.columns.values[6] = 'LGA_state'
local_gov.columns.values[1] = 'LGA_state'
poverty_rate.columns.values[4] = 'State'
poverty_rate.columns.values[2] = 'State'

poverty_rate
local_gov
```

```python
poverty_rate.columns
local_gov.columns
```

```python
#Marge the two datasets together to form a single table.
merged_data = pd.merge(local_gov, poverty_rate, on="SNo", how="inner")
merged_data
```

```python
#Remove any redundant columns
merged_cleaned = merged_data.reset_index(drop=True)
merged_cleaned = merged_data.drop('ISO', axis=1)
merged_cleaned = merged_data.drop('State_y', axis=1)
merged_data.columns.values[2] = 'State'
merged_cleaned = merged_cleaned.set_index('SNo') 

merged_cleaned
```

```python
#Clean the dataset to remove every null values
merged_cleaned = merged_cleaned.dropna(axis=1)
merged_cleaned
```

```python
#Using either groupby or pivort_table calculate how population affects poverty rate of each local government
grouped = merged_cleaned.groupby('LGA_state_x').agg(
    population_avg= ('2014_Popul', 'mean'),
    poverty_rate_avg=('Poverty_Ra', 'mean')
).reset_index()

print(grouped)
```

```python
#Arrange each local goverment according to the order of their poverty rate
grouped_sorted = grouped.sort_values(by='poverty_rate_avg', ascending=True) 

print(grouped_sorted)
```
